```python
import xarray as xr
import pandas as pd
```

```python
# create metadata csv file from netcdf with power and metadata
sme_sites = xr.open_dataset('/mnt/disks/sukh-pvnet-research-1/pvnet_smartest_energy/pv_sites/sme_all.nc')

df = sme_sites.to_dataframe()
df = df.reset_index()


dfs = []

for id in set(df['pv_id']):
    site_df = df[df["pv_id"]==id].drop_duplicates(subset=['latitude', 'longitude'], keep='first')
    dfs.append(site_df)
all_sites_meta = pd.concat(dfs)
all_sites_meta.reset_index(inplace=True)

all_sites_meta['orientation'] = 180.0
all_sites_meta['tilt'] = 35.0
all_sites_meta = all_sites_meta.drop(columns=['index', 'ts', 'power', 'StarkMeter'])

all_sites_meta = all_sites_meta.rename(columns = {"pv_id" :"system_id"})


# all_sites_meta.to_csv('/mnt/disks/sukh-pvnet-research-1/pvnet_smartest_energy/pv_sites/sme_all_metadata.csv' ,index=False)
```
